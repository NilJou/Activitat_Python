```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EXERCICI AMB NUMPY

# Exercici 1
x = np.array([1, 2, 3, 4, 5])
y = np.array([6, 7, 8, 9, 10])

xy = np.add(x, y)

# Calcule mal suma, resta, multiplicació i divisió
print("Suma: ", xy)
print("Resta: ", np.subtract(x, y))
print("Multiplicació: ", np.multiply(x, y))
print("Divisio: ", np.divide(x, y))

# Calculem la mitjana, la suma total i la desviacio estandard
print(np.average(xy))
print(np.sum(xy))
print(np.std(xy))
```

```python
# Exercici 2
df = pd.read_csv('fitxer.csv')

filtre = df[df['Edat'] > 25]


df2 = df.copy()
df2.drop('Nom', axis='columns',inplace=True)
df2.drop('Gènere', axis='columns',inplace=True)
df2.drop('Ciutat', axis='columns',inplace=True)
sort = df2[df2['Edat'] > 25].count()

print(filtre.sort_values('Edat'))
print('Nombre de registres amb edat superior a 25: ', sort)
```

```python
# EXERCICI AMB MATPLOTLIB

# Exercici 1
max_temp = [17.24, 17.18, 19.44, 21.18, 19.36, 19.78, 19.26, 22.48, 17.53, 20.08]
dies = [f"Dia {i+1}" for i in range(10)]


plt.figure(figsize=(10, 4))
plt.plot(dies, max_temp, marker = 'o', linestyle = '--', label = 'Maxima Temp')
plt.legend()
plt.xlabel('Dies')
plt.ylabel('Temp ºC')
plt.grid(True)
```

```python
# Exercici 2
dfVendes = pd.DataFrame({
    "Producte": ["Ordinador", "Teclat", "Ratolí", "Pantalla", "Impressora", "Portàtil", "Auriculars", "Altaveus", "Tablet", "Smartphone"],
    "Ciutat": ["Barcelona", "Madrid", "València", "Sevilla", "Bilbao", "Barcelona", "Madrid", "València", "Sevilla", "Bilbao"],
    "Quantitat": [5, 10, 8, 3, 6, 2, 7, 5, 9, 4],
    "Preu": [800, 30, 25, 150, 200, 1000, 50, 45, 250, 700]
})

dfVendes['Vendes'] = dfVendes['Quantitat'] * dfVendes['Preu']

vendesTotals = dfVendes.groupby('Producte')['Vendes'].sum()

# Crear el gràfic de barres
plt.figure(figsize=(10, 4))
vendesTotals.sort_values().plot(kind='bar')
plt.xlabel("Producte")
plt.ylabel("Vendes Totals")
plt.tight_layout()
```

```python
# Exercici 3
mesos = np.arange(1, 13)

A = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
B = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
C = [8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52]

plt.figure(figsize=(10, 4))
plt.fill_between(mesos, A, color='#569b39', alpha=0.3, label='Producte A')
plt.fill_between(mesos, B, color='#9b3887', alpha=0.3, label='Producte B')
plt.fill_between(mesos, C, color='#4c389b', alpha=0.3, label='Producte C')
plt.xlabel('Mesos')
plt.ylabel('Creixement')
plt.legend();
```
