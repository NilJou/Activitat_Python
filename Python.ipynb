```python
# Exercici 8
cadena = "&#Pr@j3ctE#D4M#&"
digit = 0
lletra = 0
simbol = 0

for i in cadena:
    if i.isalpha():
        lletra += 1
    elif i.isdigit():
        digit += 1
    else:
        simbol += 1
        
print("Lletes: ", lletra)
print("Digits: ", digit)
print("Simbols: ", simbol)
```
